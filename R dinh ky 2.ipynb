#### 1. Cho dữ liệu mtcars cho sẵn trong R datasets. Đọc dữ liệu. Hiển thị thông tin một số dòng đầu của dữ liệu. In thông tin thống kê chung.

#### 2. Tạo sub_data chỉ có các thông tin mpg, hp, wt, gear

```r
sub_data = select(mtcars, mpg, hp, wt, gear)
head(sub_data)
```

#### 3. Tạp cột mới km_per_litre trong sub_data theo công thức: 1 mpg = 0.425 km/l

```r
sub_data_unique = distinct(mtcars)
head(mtcars)
```

#### 5. In sub_data_unique với giá trị được sắp xếp giảm dần theo mpg, tăng dần theo hp

#### 6. Lọc các xe mà trong tên xe có 'Mazda' trong sub_data_unique. Có bao nhiêu xe thỏa điều kiện này?

#### 7. Trong sub_data_unique, cho biết giá trị mpg lớn nhất? Có những dòng dữ liệu nào có mpg = mpg lớn nhất, in kết quả.

```r
print(max(sub_data_unique$mpg))
filter(sub_data_unique, mpg == max(sub_data_unique$mpg))
```

#### 8. Lọc các dữ liệu thỏa các điểu kiện sau: mpg >= 25 và cyl=4 và gear=4

```r
head(filter(mtcars, mpg >= 25, cyl == 4, gear == 4))
```

#### 9. Thực hiện việc thống kê theo cyl, từ đó cho biết số lượng từng nhóm, cyl, hp lớn nhất, mpg nhỏ nhất

```r
grouped_data = group_by(mtcars, cyl)

summary_count = summarise(grouped_data, cyl_count = n(), max_hp = max(hp), min_mpg = min(mpg))
# summary_max_hp = summarise(grouped_data, max_hp = max(hp))
# summary_min_mpg = summarise(grouped_data, min_mpg = min(mpg))

summary_count

for (i in 1:3) {
  print(paste('cyl =', summary_count$cyl[i],'so luong la:', summary_count$cyl_count[i],
              "HP lon nhat la:", summary_count$max_hp[i],
              "MPG nho nhat la:", summary_count$min_mpg[i]))
}
```

#### 10. Trong sub_data, cho biết xe nào có km_per_litre lớn nhất, nhỏ nhất?

#### câu 1: Đọc dữ liệu, hiển thị thông tin chung của dữ liệu : head(), tail(), str(), summary()

```r
head(rent)
tail(rent)
str(rent)
summary(rent)
```

```r
# noi chuoi bang paste
print(paste('number of rows: ', ncol(rent) - 1))
print(paste('number of columns: ', nrow(rent)))
```

#### câu 3: Cho biết 5 loại rooms xuất hiện nhiều nhất trong dataset (5 loại phòng được chọn thuê nhiều nhất)

```r
x = arrange(
    summarise(
        group_by(rent, rooms), # nhom df rent voi rooms
        room_count = n() # ham chuc nang giong count nhung khong co tham so
    ),
    desc(room_count)
)
head(x, 5)
```

#### câu 4: Cho biết các cột có dữ liệu bị thiếu (na). Mỗi cột thiếu bao nhiêu giá trị? Tỷ lệ thiếu (lấy 2 số lẻ)?

```r
# tao cot, dung ham đem so luong 'na'
count_missings = colSums(is.na(rent))

# tao cot, dung ham round de lam tron 2 chu so, dung ham nrow tinh tong so luong dong
missing_percents = round(count_missings / nrow(rent) * 100, 2)

# tao bang moi
missing_info = data.frame(
  Column = names(count_missings),
  count_missings = count_missings,
  missing_percents = missing_percents
)

# loc gia tri < 0, sap xep giam dan
missing_info = filter(missing_info, count_missings > 0)
arrange(missing_info, desc(count_missings))
```

#### câu 5: Xóa bỏ các cột property.tax, fire.insurance, total. In head() để xem kết quả.

```r
rent = select(rent, -property.tax, -fire.insurance, -total)
head(rent)
```

#### câu 6: Tạo cột rent_amount từ cột rent.amount. Sau đó, bỏ ký tự "R$", và ',' ở rent_amount và đổi dữ liệu của cột này sang kiểu số. In head() để xem kết quả.

```r
# RegEx [^0-9.]
# gsub
rent = mutate(rent, rent_amount = as.numeric(gsub('[^0-9.]','', rent.amount)))
head(rent)
```

#### câu 7: Cho biết bao nhiêu nhà có nội thất, bao nhiêu nhà không có nội thất? Trong nhà có nội thất bao nhiêu nhà cho phép nuôi thú, bao nhiêu nhà không? Trong nhà không có nội thất, bao nhiêu nhà cho phép nuôi thú, bao nhiêu nhà không?

```r
fur = group_by(rent, furniture)
fur_cnt = summarise(fur, counts = n())
fur_cnt

fur_animal = group_by(rent, furniture, animal)
fur_animal_cnt = summarise(fur_animal, counts = n())
fur_animal_cnt
```

#### 8. Vẽ boxplot của cột rent_amount theo furniture và nhận xét.

#### 9. Cột area có các giá trị na, hãy thay thế các giá trị na bằng giá trị median.

```r
# tinh median cua area, bo qua gia tri na
median_area = median(rent$area, na.rm = TRUE)

# thay the gia tri na
rent$area = ifelse(is.na(rent$area), median_area, rent$area)
head(rent)
```

#### 10. Vẽ boxplot của cột area và nhận xét. Cột area có outlier không? Nếu có cho biết tổng số mẫu outlier.

```r
summary(rent$area)
boxplot(rent$area, ylim = c(0, 1000))
```

#### 11. Vẽ biểu đồ thể hiện mối liên hệ của area và rent_amount

```r
summary(rent$rent_amount)
summary(rent$area)
# do min va max cua x, y do lech lon hon nhieu so voi cac gia tri khac nen em dung tu phan vi
plot(x = rent$rent_amount, y = rent$area,
     xlab =' gia thanh', ylab = 'dien tich',
     main = 'moi quan he giua gia thanh va dien tich',
     xlim = c(1800, 5952), ylim = c(58, 200)
    )
```

#### 12. Vẽ pie chart thể hiện \% giữa 1 và 0 của cột city

```r
group = group_by(rent, city)
city_cnt = summarise(group, city_count =  n())
head(city_cnt)

city_percent = round(100 * city_cnt$city_count/sum(city_cnt$city_count),2)
pie(city_cnt$city_count, labels = city_percent,
    main ='phan tram cua nha o thanh pho',
    col= rainbow(length(city_cnt$city_count)))
```

#### 13. Cho biết rent_amount lớn nhất và bé nhất? Liệt kê những căn nhà có rent_amount lớn nhất và bé nhất.
