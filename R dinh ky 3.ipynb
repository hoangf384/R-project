## câu 3: Cho biết 5 loại rooms xuất hiện nhiều nhất trong dataset (5 loại phòng được chọn thuê nhiều nhất)

```r
x = arrange(
    summarise(
        group_by(rent, rooms), # nhom df rent voi rooms
        room_count = n() # ham chuc nang giong count nhung khong co tham so
    ),
    desc(room_count)
)
head(x, 5)
```

## câu 4: Cho biết các cột có dữ liệu bị thiếu (na). Mỗi cột thiếu bao nhiêu giá trị? Tỷ lệ thiếu (lấy 2 số lẻ)?

```r
# tao cot, dung ham đem so luong 'na'
count_missings = colSums(is.na(rent))

# tao cot, dung ham round de lam tron 2 chu so, dung ham nrow tinh tong so luong dong
missing_percents = round(count_missings / nrow(rent) * 100, 2)

# tao bang moi
missing_info = data.frame(
  Column = names(count_missings),
  count_missings = count_missings,
  missing_percents = missing_percents
)

# loc gia tri < 0, sap xep giam dan
missing_info = filter(missing_info, count_missings > 0)
arrange(missing_info, desc(count_missings))
```

## câu 6: Tạo cột rent_amount từ cột rent.amount. Sau đó, bỏ ký tự "R$", và ',' ở rent_amount và đổi dữ liệu của cột này sang kiểu số. In head() để xem kết quả.

```r
# RegEx [^0-9.]
# gsub
rent = mutate(rent, rent_amount = as.numeric(gsub('[^0-9.]','', rent.amount)))
head(rent)
```
